```python
#Bibliotecas necessárias
import pandas as pd
import numpy as np
```

```python
#Leitura do CSV
df = pd.read_csv("unis.csv")
```

```python
#Conversão das notas, de string para float
df['Nota'] = df['Nota'].str.replace(',','.').astype(float)
```

```python
#1) Qual foi o aluno com a maior quantidade de aplicações?
alunoapl = df['Aluno'].value_counts()
print(alunoapl)
print("\nAluno:",alunoapl.idxmax())
print("Nº de Aplicações:",alunoapl.max())
```

```python
#2) Qual foi o aluno com a maior quantidade de aprovações?
alunoapr = df[df.Resultado == 'Aceita'].Aluno.value_counts()
print(alunoapr)
print("\nAluno:",alunoapr.idxmax())
print("Nº de Aprovações:",alunoapr.max())
```

```python
#3) Qual foi o aluno que escolheu a maior quantidade de Universidades?
alunoesc = df[df.Resultado == 'Escolhida'].Aluno.value_counts()
print(alunoesc)
print("\nAluno:",alunoesc.idxmax())
print("Nº de Universidades Escolhidas:",alunoesc.max())
```

```python
#4) Qual foi o aluno que teve o maior percentual de aproveitamento e qual foi o percentual? (Aprovações X Total de aplicações)
percent = alunoapr/alunoapl
print("\nAluno:", percent.idxmax())
print("Aproveitamento:", round(percent.max() * 100,2),"%")
```

```python
#5) Qual foi a universidade com a maior quantidade de aplicações?
uniapl = df['Universidade'].value_counts()
print(uniapl)
print("\nUniversidade:",uniapl.idxmax())
print("Nº de Aplicações:",uniapl.max())
```

```python
#6) Qual ou quais foram as universidade que tiveram a maior nota necessária para aprovação?
grade = df[(df['Nota'] == df['Nota'].max()) & (df.Resultado == 'Aceita')]
print(grade[['Universidade','Resultado','Nota']])
print("\nUniversidades:",grade['Universidade'].unique())
```
